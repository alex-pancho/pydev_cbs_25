## 4. –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è —à–∞–±–ª–æ–Ω—É —É view

### –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è `render()`

**blog/views.py**

```python
from django.shortcuts import render

def blog_index(request):
    return render(request, 'blog/index.html')
```

## 5. –ü–µ—Ä–µ–¥–∞—á–∞ –¥–∞–Ω–∏—Ö —É —à–∞–±–ª–æ–Ω (context)

### View

```python
def blog_index(request):
    context = {
        'title': 'My Blog',
        'posts_count': 5
    }
    return render(request, 'blog/index.html', context)
```

### Template

```html
<h1>{{ title }}</h1>
<p>Total posts: {{ posts_count }}</p>
```

üìå `{{ }}` ‚Äî –≤–∏–≤—ñ–¥ –∑–º—ñ–Ω–Ω–∏—Ö

---

## 6. –£–º–æ–≤–∏ –≤ —à–∞–±–ª–æ–Ω–∞—Ö (if)

```html
{% if posts_count > 0 %}
    <p>Posts available</p>
{% else %}
    <p>No posts yet</p>
{% endif %}
```

## 7. –¶–∏–∫–ª–∏ (for)

### View

```python
def blog_index(request):
    posts = ['First post', 'Second post', 'Third post']
    return render(request, 'blog/index.html', {'posts': posts})
```

### Template

```html
<ul>
{% for post in posts %}
    <li>{{ post }}</li>
{% empty %}
    <li>No posts</li>
{% endfor %}
</ul>
```

## 9. –ù–∞—Å–ª—ñ–¥—É–≤–∞–Ω–Ω—è —à–∞–±–ª–æ–Ω—ñ–≤ (base.html)

### base.html

```html
<!DOCTYPE html>
<html>
<head>
    <title>{% block title %}Site{% endblock %}</title>
</head>
<body>
    <header>Header</header>

    {% block content %}
    {% endblock %}

    <footer>Footer</footer>
</body>
</html>
```

### index.html

```html
{% extends 'blog/base.html' %}

{% block title %}Blog{% endblock %}

{% block content %}
<h1>Blog page</h1>
{% endblock %}
```
